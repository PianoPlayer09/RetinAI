```python
def get_class_from_filename(name):
  classes = name.split('_')[0]
  return classes
```

```python
import os
import random
import shutil


train_path = '/content/images/train'
val_path = '/content/images/val'
test_path = '/content/images/test'


images_path = "/content/preprocessed_images"
images = os.listdir(images_path)
random.shuffle(images)


if not os.path.exists(train_path):
 os.makedirs(train_path)
if not os.path.exists(val_path):
 os.makedirs(val_path)
if not os.path.exists(test_path):
 os.makedirs(test_path)


train_split = int(0.8*len(images))
val_split = int(0.1*len(images))
test_split = int(0.1*len(images)) # not needed


test = train_split+val_split


train_set = images[:train_split]
val_set = images[train_split:test]
test_set = images[test:]

classes = ['N', 'D', 'G', 'C', 'A', 'H', 'P', 'O']

for base_path in [train_path, val_path, test_path]:
  for cls in classes:
    class_folder = os.path.join(base_path, cls)
    if not os.path.exists(class_folder):
      os.makedirs(class_folder)


for image in train_set:
 cls = get_class_from_filename(image)
 src_path = os.path.join(images_path, image)
 dest_path = os.path.join(train_path,cls, image)
 shutil.move(src_path, dest_path)
for image in val_set:
 cls = get_class_from_filename(image)
 src_path = os.path.join(images_path, image)
 dest_path = os.path.join(val_path,cls, image)
 shutil.move(src_path, dest_path)
for image in test_set:
 cls = get_class_from_filename(image)
 src_path = os.path.join(images_path, image)
 dest_path = os.path.join(test_path,cls, image)
 shutil.move(src_path, dest_path)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torchvision
import torchvision.transforms as transforms


transform = transforms.Compose([
   transforms.Resize((222, 222)),
   transforms.ToTensor(),
   transforms.Normalize(mean=[0.5], std=[0.5])
])


batch_size = 32
trainset = torchvision.datasets.ImageFolder(root='/content/images/train', transform=transform)
trainLoader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle = True)
testset = torchvision.datasets.ImageFolder(root='/content/images/test', transform=transform)
testLoader = torch.utils.data.DataLoader(testset, batch_size=batch_size, shuffle = True)
```

```python
# visualize images
def imshow(img):
  img = img/2 + 0.5
  npimg = img.numpy()
  plt.show()

dataiter = iter(trainLoader)
images, labels = next(dataiter)
imshow(torchvision.utils.make_grid(images, nrow=2))


print(f"Image batch shape: {images.shape}")  # Should be [batch_size, channels, height, width]
print(f"Label batch shape: {labels.shape}")
```

```python
import torch.nn as nn

CLASSES = ['Normal', 'Diabetes', 'Cataract', 'Age Related Macular Degeneration', 'Glaucoma', 'Hypertension', 'Pathological Myopia', 'Other Disease']
NUM_CLASSES = len(CLASSES)

class EyeMultiClassClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = nn.Conv2d(3, 4, 3) # 16, 220, 220
    self.pool = nn.MaxPool2d(2, 2) # 16, 110, 110
    self.conv2 = nn.Conv2d(4, 8, 3) # 32, 108, 108, # out after pooling: 32, 54, 54
    self.conv3 = nn.Conv2d(8, 12, 3) # 32, 52, 52, # out after pooling: 32, 26, 26
    self.conv4 = nn.Conv2d(12, 16, 3) # 32, 24, 24, # out after pooling: 16, 12, 12
    self.lin1 = nn.Linear(16*12*12, 256)
    self.lin2 = nn.Linear(256, 128)
    self.lin3 = nn.Linear(128, 64)
    self.lin4 = nn.Linear(64, NUM_CLASSES)
    self.relu = nn.ReLU()
    self.log_softmax = nn.LogSoftmax(dim=1)

  def forward(self, x):
    x = self.conv1(x)
    x = self.relu(x)
    x = self.pool(x)
    x = self.conv2(x)
    x = self.relu(x)
    x = self.pool(x)
    x = self.conv3(x)
    x = self.relu(x)
    x = self.pool(x)
    x = self.conv4(x)
    x = self.relu(x)
    x = self.pool(x)
    x = torch.flatten(x, 1)
    x = self.lin1(x)
    x = self.relu(x)
    x = self.lin2(x)
    x = self.relu(x)
    x = self.lin3(x)
    x = self.relu(x)
    x = self.lin4(x)
    x = self.log_softmax(x)
    return x

model = EyeMultiClassClassifier()
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
# %% training
NUM_EPOCHS = 5
for epoch in range(NUM_EPOCHS):
    for i, data in enumerate(trainLoader, 0):
        inputs, labels = data
        # TODO: define training loop
        optimizer.zero_grad()

        outputs = model(inputs)

        loss = loss_fn(outputs, labels)

        loss.backward()

        optimizer.step()


    print(f'Epoch {epoch}/{NUM_EPOCHS}, Loss: {loss.item():.4f}')

```

```python
# %% test
y_test = []
y_test_hat = []
for i, data in enumerate(testLoader, 0):
    inputs, y_test_temp = data
    with torch.no_grad():
        y_test_hat_temp = model(inputs).round()

    y_test.extend(y_test_temp.numpy())
    y_test_hat.extend(y_test_hat_temp.numpy())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np

acc = accuracy_score(y_test, np.argmax(y_test_hat, axis=1))
print(f'Accuracy: {acc*100:.2f} %')
```

```python
confusion_matrix(y_test, np.argmax(y_test_hat, axis=1))
```

```python
import torch

# Save checkpoint
checkpoint = {
    'epoch': epoch,
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'loss': loss,
}
torch.save(checkpoint, 'checkpoint_epoch5.pth')
```
